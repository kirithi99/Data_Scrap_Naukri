```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://www.naukri.com/description-jobs")
```

```python
soup=BeautifulSoup(driver.page_source,"html.parser")
```

```python
soup
```

```python
title=soup.find_all('div',class_="row1")
```

```python
title_list=[]
for i in title:
    title_list.append(i.find('a',class_="title").text)
```

```python
title_list
```

```python
company=soup.find_all('div',class_='row2')
```

```python
company_list=[]
for i in company:
    company_list.append(i.find('span',class_="comp-dtls-wrap").text)
```

```python
company_list
```

```python
experience=soup.find_all('div',class_="row3")
```

```python
exp_list=[]
for i in experience:
    exp_list.append(i.find('span',class_='exp-wrap').text)
```

```python
exp_list
```

```python
job_location=soup.find_all('div',class_='row3')
```

```python
job_loc=[]
for i in job_location:
    job_loc.append(i.find('span',class_='loc-wrap ver-line').text)
```

```python
job_loc
```

```python
job_details=[]
for i in soup.find_all('div',class_="srp-jobtuple-wrapper"):
    title=i.find('a',class_="title").text
    company=i.find('span',class_="comp-dtls-wrap").text
    experience=i.find('span',class_='exp-wrap').text
    job_loc=i.find('span',class_='loc-wrap ver-line').text
    job_details.append({
         "JOB TITLE" : title,
         "COMPANY" : company,
          "EXPERIENCE":experience,
          "JOB LOCATION" :job_loc
    })
```

```python
job_details
```

```python
import pandas as pd
#creating a dataframe
df = pd.DataFrame(job_details)
#save the data to a csv file
df.to_csv('Job_Details_From_Naukri.csv',index=False,encoding='utf-8')
print(df)
```
